```python
def choose_challange_by_Gender(gender,challange):
    CurrDf=df[df['Challange_Name']==challange]
    DfGender=CurrDf[CurrDf['User_Gender'] ==gender]
    return DfGender.shape[0]

choose_challange_by_Gender("Male", "Min Window Substring")    
```

```python
def create_df_of_currChallange(chall):
    currDF=df[df['Challange_Name']==chall]
    return currDF 
```

```python
def Table_of_Data_orderBy_GenderAndChallange():
    data = {'Gender': ['Male', 'Female'],  }
    listChallange=[]
    listChallange=df['Challange_Name'].value_counts()[:].index.tolist() 
    finalDF = pd.DataFrame(data )

    for i in listChallange:
        currDF=create_df_of_currChallange(i)
        maleCount=currDF[currDF['User_Gender']=='Male'].shape[0] 
        femaleCount=currDF[currDF['User_Gender']=='Female'].shape[0]
        finalDF.insert( 1 , i, [maleCount , femaleCount], True) 
        
    TotalMaleCount=df[df['User_Gender']=='Male'].shape[0]
    TotalFemaleCount=df[df['User_Gender']=='Female'].shape[0]  
    finalDF.insert( 12 , "Total For Gender", [TotalMaleCount , TotalFemaleCount], True)
    return finalDF
Table_of_Data_orderBy_GenderAndChallange()
```

```python
def show_pie_by_Challange(chall):
    h=Table_of_Data_orderBy_GenderAndChallange()
    mycolors=['#1d37d2','#f95ade' ] ;labels =    'Male', 'Female'
    plt.title(chall)
    return plt.pie(h[chall],colors=mycolors ,labels= labels ,autopct='%1.1f%%',shadow=True )
# show_pie_by_Challange("First Reverse")
```

```python
df = pd.read_csv(r'C:\Users\Public\python\ANACONDA\DATA SCIENCE\project with noam\project_data_3000.csv' )
def Find_identical_Couples_language(lang1,lang2):
    currDF=dfZero.loc[:, [  lang1,lang2] ]
    currDF=currDF.dropna( )
    return  currDF.shape[0]- 60

Find_identical_Couples_language('C','Python')
```

```python
lang_list=df['Programming_Lang'].unique()
TotalCouplesDF = pd.DataFrame(np.zeros((len(lang_list),len(lang_list))))
TotalCouplesDF.columns = TotalCouplesDF.index= lang_list  
for i in lang_list:
    for j in lang_list:
        if(i==j):
            TotalCouplesDF[i][j]=None
        else:
            TotalCouplesDF[i][j]=Find_identical_Couples_language(i,j)

TotalCouplesDF   
```

```python
#this func replace in the df all the 0 and 1 values to become null
def replacing_df_into_null(dataframe):
    dataframe=dataframe.replace(1,np.nan)
    # barDf=barDf.replace(2,np.nan)
    dataframe=dataframe.replace(0,np.nan)
    dataframe=dataframe.dropna( axis =0,thresh=1)
    dataframe=dataframe.dropna( axis =1,thresh=1)
    dataframe=dataframe .replace(np.nan,0)
    return dataframe    
```

```python
#this func get dictonary contain all the data and create bar graph
def create_bar_graph(dict):
    languages = (dict.keys()) 
    values = (dict.values())
    fig = plt.figure(figsize = (10, 5)) 
    for a,b in zip(languages, values):
        plt.text(a, b, str(b))
    plt.bar(languages, values, color ='blue',  width = 0.1) 
```

```python
#this function is the main function- its calculate the relationship_between_languages and show it in bar graph
def relationship_between_languages():
    lang_list=df['Programming_Lang'].unique()
    barDf = pd.DataFrame(TotalCouplesDF ) 
    barDf=replacing_df_into_null(barDf)
    dict={}
    lang_list=barDf.columns.values.tolist()
    for i in lang_list:
        for j in lang_list:
            if(barDf[i][j]>0):
                dict[i +"\n And "+ j] = barDf[i][j]
        lang_list.remove(i)
    return create_bar_graph(dict)
    
relationship_between_languages()
```
